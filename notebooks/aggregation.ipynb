```python
import pandas as pd

df = pd.read_csv("F:\OneDrive\psl-data-analysis\data\psl_cleaned.csv")
```

```python
team_runs = df.groupby("batting_team")["runs_off_bat"].sum()

team_runs
```

```python
df.groupby("batting_team")["runs_off_bat"].sum().sort_values(ascending=False)
df  # total runs by each team
```

```python
df.groupby("bowler")["wicket_type"].count().sort_values(ascending=False)
df  #Total wickets taken by each player (bowler)
```

```python
df.groupby("striker")["runs_off_bat"].sum().sort_values(ascending=False)
df  #Total wickets taken by each bowler
```

```python
df.groupby("striker")["ball"].count().sort_values(ascending=False)
df  #Total balls faced by each striker
```

```python
df.groupby("bowling_team")["extras"].sum().sort_values(ascending=False)
df  #Total extras given by each team
```

```python
df.groupby("batting_team")["runs_off_bat"].mean()
df  #Average runs per innings by team
```
